```python
import pandas as pd
```

```python
hf = pd.read_csv('hurricane_frequency.csv')
hf.head()
```

```python
summed_hf = hf[['Tropical Storms', 'Hurricanes', 'Major Hurricanes']].groupby((hf.Year//10)*10).sum().reset_index()
summed_hf.head()
```

```python
mean_hf = hf[['Tropical Storms', 'Hurricanes', 'Major Hurricanes']].groupby((hf.Year//10)*10).mean()
mean_hf
```

```python
mean_hf.to_csv('hf_decade_avg.csv')
```
